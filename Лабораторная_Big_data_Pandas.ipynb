```python
df_state = df.groupby('state')[['churned']].mean()
```

```python
df_state = df[['state', 'churned']].where(df['churned']==True).groupby('state').count().rename(columns={'churned': 'churned_count'})
```

```python
df_state['count'] = df[['state', 'churned']].groupby('state').count().rename(columns={'churned': 'count'})['count']
```

```python
df_state.head()
```

```python
df_state['part_churned'] = df_state['churned_count'] / df_state['count']
```

```python
df_state.head()
```

```python
df_state = df_state.join(df[['state', 'total_day_minutes', 'total_day_charge']].groupby('state').sum(), on='state')
```

```python
df_state['avg_day_charge'] = df_state['total_day_charge'] / df_state['total_day_minutes']
```

```python
df_state[['churned', 'avg_day_charge']].head()
```

```python
df = pd.get_dummies(df, columns=['state'])
```

```python
df.info()
```

```python
df = df.drop(columns = ['area_code', 'phone_number', 'international_plan', 'voice_mail_plan', 'total_minutes'])
```

```python
df.head()
```

```python
x, y = df.drop(columns='churned'), df['churned']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train, model.predict(x_train)))
print(accuracy_score(y_test, model.predict(x_test)))
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model_tree = DecisionTreeClassifier()
model_tree.fit(x_train, y_train)
print(accuracy_score(y_train, model_tree.predict(x_train)))
print(accuracy_score(y_test, model_tree.predict(x_test)))
```
