## Загрузите данные при помощи spark.read.csv из приложенного файла

```
df = spark.read.csv('u.data.csv', inferSchema=True, header=False, sep='\t')
```

```
df.show(5)
```

```
df = df.withColumnRenamed('_c0', 'user_id')
df = df.withColumnRenamed('_c1', 'film_id')
df = df.withColumnRenamed('_c2', 'rating')
df = df.withColumnRenamed('_c3', 'timestamp')
```

```
item = spark.read.csv('u.item.csv', inferSchema=True, header=False, sep='|')
```

```
item.show()
```

```
oldColumns = item.schema.names
newColumns = ['movie_id', 'movie_title', 'release_date', 'video_release_date',
              'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation',
              'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
              'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi',
              'Thriller', 'War', 'Western']
for i in range(len(newColumns)):
  item = item.withColumnRenamed(oldColumns[i], newColumns[i])
```

```
item.show()
```

## Рассчитаем среднюю оценку кажного фильма и каждого жанра

```
avg_ratings = df.groupby('film_id').mean('rating')
avg_ratings.show()
```

```
df = df.join(item, (df['film_id']==item['movie_id']), how='left')
df.show()
```

Выведем среднюю оценку фильма с названием

```
df.groupby('film_id', 'movie_title').mean('rating').show()
```

```
Action_rat = df.where(df['Action'] == 1).groupby('Action').mean('rating')
```

```
Action_rat.first()['avg(rating)']
```

```
genres = ['Action', 'Adventure', 'Animation',
              'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
              'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi',
              'Thriller', 'War', 'Western']
gen_rat =[]
```

```
for i in range(len(genres)):
  gen_rat.append(df.where(df[genres[i]] == 1).groupby(genres[i]).mean('rating').first()['avg(rating)'])
```

```
print('Средний рейтинг по жанрам:')
for i in range(len(genres)):
  print(genres[i], '-', gen_rat[i])
```

## Выведем 5 самых популярных и самых непопулярных фильмов по количеству оценок.

```
df.groupby('film_id', 'movie_title').count().sort('count').show(5)
```

```
df_r_count = df.groupby('film_id', 'movie_title').count()
df_r_count.where(df_r_count['count']==1).count()
```

Есть 141 фильм, у которого только одна оценка. Мы не знаем, какие из них включить в топ-5, поэтому включим их все.

```
df_r_count_low = df_r_count.where(df_r_count['count']==1)
```

```
df_r_count_low.show(141)
```

Чтобы создать датафрейм с топ-5 самых оцениваемых фильмов, отсортируем и проиндексируем датафрейм, а потом оставим первые пять строчек.

```
from pyspark.sql.functions import monotonically_increasing_id
df_r_count_high = df.groupby('film_id', 'movie_title').count().sort('count', ascending=False)
df_r_count_high = df_r_count_high.withColumn("index", monotonically_increasing_id())
df_r_count_high = df_r_count_high.where(df_r_count_high['index'] < 5)
```

```
df_r_count_high.show()
```
